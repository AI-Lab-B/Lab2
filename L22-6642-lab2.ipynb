```python
with open('genome.txt', 'r') as file:

    lines = [line.strip() for line in file]


print(lines)


nucleotide = list(lines[0])
print(nucleotide)
genome_length = len(nucleotide)
print(genome_length)
t= np.linspace(0, 4 * np.pi, genome_length)

x = np.cos(t)
y = np.sin(t)
z = np.linspace(0, 5, genome_length)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
coordinates = np.column_stack((x, y, z))

colors = {'A': 'red', 'T': 'blue', 'C': 'green', 'G': 'orange'}

for i, nucleotide in enumerate(nucleotide):
    ax.scatter(coordinates[i, 0], coordinates[i, 1], coordinates[i, 2],
    color=colors[nucleotide], marker='o')

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Genome Visualization")
plt.show()

```

```python
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt

iris = load_iris()

X = np.array(iris.data)
Y = np.array(iris.target)

for i in range(4):
  means = np.mean(X[:,i])
  medians = np.median(X[:,i])
  stds = np.std(X[:,i])
  mins = np.min(X[:,i])
  maxs = np.max(X[:,i])
  print("Feature Means:", means)
  print("Feature Medians:", medians)
  print("Feature Standard Deviations:", stds)
  print("Feature Minimums:",mins)
  print("Feature Maximums:", maxs)

sepal_length_width = X[:,:2]
print("Sepal Length and Width:\n", sepal_length_width)


plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=Y, cmap='viridis')
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Scatter Plot: Sepal Length vs Sepal Width")
plt.colorbar(label="Species")
plt.show()


plt.figure(figsize=(8, 6))
plt.hist(X[:, 0], bins=20, color='skyblue', edgecolor='black')
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.title("Histogram: Distribution of Sepal Length")
plt.show()


plt.figure(figsize=(8, 6))
plt.plot(X[:, 2], X[:, 3], marker='o', linestyle='-', color='coral')
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.title("Line Plot: Petal Length vs Petal Width")
plt.show()

```
