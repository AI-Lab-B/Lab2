```python
#Q2

genome = []
with open("genome.txt", 'r') as file:
    for line in file:
        genome.append(line.strip())

genome = ''.join(genome)
length = len(genome)

print("The genome is: ", genome)
print("The length of the genome is: ", length)

t = np.linspace(0, 4*np.pi, length)
x = np.cos(t)
y = np.sin(t)
z = np.linspace(0, 5, length)

coordinates = np.column_stack((x, y, z))

#assigning colour to each molecule

colors = {
    'A': 'red',
    'T': 'yellow',
    'C': 'cyan',
    'G': 'green'
}

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
for i in range(length):
    ax.scatter(coordinates[i, 0], coordinates[i, 1], coordinates[i, 2], c=colors[genome[i]], marker='o', s=50)

plt.show()


```

```python
#Q3 

import matplotlib.image as mpimg

img = mpimg.imread('Dr Doom.jpg')

array = np.array(img)
rot_array = np.rot90(array)
flip_array = np.fliplr(array)
grayscale = np.dot(array[...,:3], [0.299, 0.587, 0.114])

plt.subplot(2,2,1)
plt.title('Dr Doom')
plt.axis('off')
plt.imshow(array)

plt.subplot(2,2,2)
plt.title('Horizontal Dr Doom')
plt.axis('off')
plt.imshow(rot_array)

plt.subplot(2,2,3)
plt.title('Doom Dr')
plt.axis('off')
plt.imshow(flip_array)

plt.subplot(2,2,4)
plt.title('DOOOMMMM')
plt.axis('off')
plt.imshow(grayscale, cmap='gray')

```

```python
#Q4

from sklearn.datasets import load_iris

iris = load_iris()

X = np.array(iris.data)

Y = np.array(iris.target)

#mean, median , standard deviation , max and min value of each feature

mean = np.mean(X, axis=0)
median = np.median(X, axis=0)
std_dev = np.std(X, axis=0)
max_val = np.max(X, axis=0)
min_val = np.min(X, axis=0)

print("Mean: ", mean)
print("Median: ", median)
print("Standard Deviation: ", std_dev)
print("Max Value: ", max_val)
print("Min Value: ", min_val)

#plots using matplotlib

sepal_length = X[:, 0]
sepal_width = X[:, 1]

plt.subplot(2,2,1)
plt.scatter(sepal_length, sepal_width)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

plt.subplot(2,2,2)
plt.hist(sepal_length, bins=20)
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')

petal_length = X[:, 2]
petal_width = X[:, 3]

plt.subplot(2,2,3)
plt.plot(petal_length, petal_width)
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')

plt.show()
```
