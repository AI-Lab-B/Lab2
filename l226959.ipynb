```python
import matplotlib.pyplot as plt
import numpy as np
import random
# Open the file in read mode
file = open("genome.txt", "r")
# Read the entire content of the file
content = file.read()
# Print the content
print(content)

# Close the file
file.close()
nucleotides=list(content)
genome_length=len(nucleotides)
print(genome_length)
# We'll use the parametric equations for a helix:
# x = cos(t), y = sin(t), z = t (or a scaled version of t)
# We want to span a range so that the helix makes a few turns.
t = np.linspace(0, 4 * np.pi, genome_length) # 4*pi gives about

x = np.cos(t)
y = np.sin(t)
z = np.linspace(0, 5, genome_length) # z increases linearly to

# Combine the coordinates into a (genome_length x 3) array
coordinates = np.column_stack((x, y, z))
# Create 3D plot
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
# Color mapping
colors = {'A': 'red', 'T': 'blue', 'C': 'green', 'G': 'orange'}
for i, nucleotide in enumerate(nucleotides):
    ax.scatter(coordinates[i, 0], coordinates[i, 1], coordinates[i, 2],color=colors[nucleotide], marker='o')
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Genome Visualization")
plt.show()
```
