```python
import random
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

with open("/content/sample_data/gnome.txt", "r") as file:
    content = file.read()


print(content)

listt = list(content)
print(listt)


genome_length = len(content)


genome = "".join(random.choice(listt) for _ in range(genome_length))

t = np.linspace(0, 4 * np.pi, genome_length)
x = np.cos(t)
y = np.sin(t)
z = np.linspace(0, 5, genome_length)


coordinates = np.column_stack((x, y, z))

figure = plt.figure(figsize=(12, 12))
ax = figure.add_subplot(111, projection='3d')


colors = {'A': 'red', 'T': 'blue', 'C': 'green', 'G': 'orange'}


for i, nucleotide in enumerate(genome):
    ax.scatter(coordinates[i, 0], coordinates[i, 1], coordinates[i, 2],
               color=colors.get(nucleotide, 'black'), marker='o')

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Genome Visualization on a Helix")


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image


image = Image.open('images.jfif')
img_array = np.array(image)

plt.imshow(img_array)
plt.title("Original Image")
plt.axis('off')
plt.show()

rotated_img = np.rot90(img_array)
flipped_img = np.fliplr(img_array)


plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(rotated_img)
plt.title("Rotated Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(flipped_img)
plt.title("Flipped Image")
plt.axis('off')

plt.show()

gray_img = np.dot(img_array[..., :3], [0.299, 0.587, 0.114])

plt.imshow(gray_img, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
plt.show()
```

```python
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt

iris = load_iris()
X = np.array(iris.data)
Y = np.array(iris.target)

feature_means = np.mean(X, axis=0)
feature_medians = np.median(X, axis=0)
feature_stds = np.std(X, axis=0)


feature_mins = np.min(X, axis=0)
feature_maxs = np.max(X, axis=0)

sepal_data = X[:, :2]


print(" Means:", feature_means)
print(" Medians:", feature_medians)
print(" Standard Deviations:", feature_stds)
print(" Minimums:", feature_mins)
print(" Maximums:", feature_maxs)

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=Y, cmap='viridis')
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Sepal Length vs. Sepal Width")
plt.colorbar(label="Species")
plt.show()


plt.figure(figsize=(8, 6))
plt.hist(X[:, 0], bins=10, color='skyblue', edgecolor='black')
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.title("Distribution of Sepal Length")
plt.show()

plt.figure(figsize=(8, 6))
plt.plot(X[:, 2], X[:, 3], marker='o', linestyle='-', color='orange')
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.title("Petal Length vs. Petal Width")
plt.show()
```
