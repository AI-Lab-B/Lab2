```python
import matplotlib.pyplot as plt
import numpy as np
import random

with open("text.txt", "r") as file:
    genome_seq = file.read()

# Generate random genome sequence
nucleotides = ["A", "T", "C", "G"]
genome_length = len(genome_seq)
genome = "".join(random.choice(nucleotides) for _ in range(genome_length))

t = np.linspace(0, 4 * np.pi, genome_length)
x = np.cos(t)
y = np.sin(t)
z = np.linspace(0, 5, genome_length)
coordinates = np.column_stack((x, y, z))

colors = {'A': 'red', 'T': 'blue', 'C': 'green', 'G': 'orange'}

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(projection='3d')

for i, nucleotide in enumerate(genome):
    ax.scatter(coordinates[i, 0], coordinates[i, 1], coordinates[i, 2],
               color=colors[nucleotide], marker='o')

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Genome Helix")
plt.show()
```

```python
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt

iris = load_iris()
X = np.array(iris.data)
Y = np.array(iris.target)

mean = np.mean(X, axis=0)
median = np.median(X, axis=0)
std_dev = np.std(X, axis=0)
min_values = np.min(X, axis=0)
max_values = np.max(X, axis=0)

sepal_features = X[:, :2]

plt.scatter(X[:, 0], X[:, 1], c=Y, cmap='viridis')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal Length vs Sepal Width')
plt.show()

plt.hist(X[:, 0], bins=20, color='blue', alpha=0.7)
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.title('Distribution of Sepal Length')
plt.show()

plt.plot(X[:, 2], X[:, 3], marker='o', linestyle='-', color='red')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Petal Length vs Petal Width')
plt.show()
```
