```python
with open('genome.txt', 'r') as file:

    lines = [line.strip() for line in file]


print(lines)


nucleotide = list(lines[0])
print(nucleotide)
genome_length = len(nucleotide)
print(genome_length)
t= np.linspace(0, 4 * np.pi, genome_length)

x = np.cos(t)
y = np.sin(t)
z = np.linspace(0, 5, genome_length)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
coordinates = np.column_stack((x, y, z))

colors = {'A': 'red', 'T': 'blue', 'C': 'green', 'G': 'orange'}

for i, nucleotide in enumerate(nucleotide):
    ax.scatter(coordinates[i, 0], coordinates[i, 1], coordinates[i, 2],
    color=colors[nucleotide], marker='o')

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Genome Visualization")
plt.show()
```

```python
from PIL import Image
from numpy import asarray
import matplotlib.pyplot as plt

# load the image and convert into
# numpy array

img = Image.open('goat.jfif')
numpydata = asarray(img)


plt.imshow(numpydata)
plt.title('Image from NumPy Array')
plt.show()


rotImg = np.rot90(numpydata)
plt.imshow(rotImg)
plt.title('Rotated Image')
plt.show()



flipImage=np.flip(numpydata,0)
plt.imshow(flipImage)
plt.title('Flipped Image')
plt.show()



gray_img = np.dot (numpydata[..., :3], [0.299, 0.587, 0.114])
plt.imshow(gray_img)
plt.title('Gray Image')
plt.show()


```

```python
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt



# Load the Iris dataset
iris = load_iris()

# Accessing the features (data) using NumPy array
X = np.array(iris.data)
Y = np.array(iris.target)


#print(X)
#print(Y)

print(X.shape[1])




for i in range(X.shape[1]):
    feature = X[:, i]
    mean = np.mean(feature)
    median = np.median(feature)
    std_dev = np.std(feature)
    min_val = np.min(feature)
    max_val = np.max(feature)

    print(f"Feature {i + 1}:")
    print(f"  Mean: {mean:}")
    print(f"  Median: {median:}")
    print(f"  Standard Deviation: {std_dev:}")
    print(f"  Minimum: {min_val:}")
    print(f"  Maximum: {max_val:}")
    print("-------------------------------------" )




len_Width=X[:,:2]
print("Length | Width (1st and second column respectively)")
print(len_Width)



plt.scatter(X[:, 0], X[:, 1], c=Y)
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Sepal Length vs. Sepal Width")

plt.show()



plt.hist(X[:, 0], bins=20, color='skyblue', edgecolor='black')
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.title("Distribution of Sepal Length")
plt.show()


plt.plot(X[:, 2], X[:, 3], marker='o', linestyle='-', color='coral')
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.title("Petal Length vs. Petal Width")
plt.show()







```
