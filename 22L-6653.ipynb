```python
import matplotlib.pyplot as plt
import numpy as np

# Using with statement to handle file opening and closing automatically
with open("genome.txt") as file:
    genome_sequence = file.read().strip()  # Strip to remove leading/trailing whitespace/newlines

# Length of the genome sequence
genome_length = len(genome_sequence)
print(f"Genome Length: {genome_length}")

# Prepare coordinates for the helix
t = np.linspace(0, 4 * np.pi, genome_length)
x = np.cos(t)
y = np.sin(t)
z = np.linspace(0, 5, genome_length)

# Color map for bases
color_map = {
    "A": "red",
    "T": "blue",
    "G": "green",
    "C": "yellow"
}

# Create figure and 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Generate coordinates and plot each point
for i, (x_val, y_val, z_val) in enumerate(zip(x, y, z)):
    base = genome_sequence[i]
    ax.scatter(x_val, y_val, z_val, c=color_map.get(base, "gray"), s=100)

# Set labels and title
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.title("Helix Structure of Genome Sequence")

# Show the plot
plt.show()
```

```python
from PIL import Image
import matplotlib.pyplot as plt
from numpy import asarray

img=Image.open("download.jpg")
image_arr=asarray(img)

# print (image_arr)

print("Simple image")
plt.imshow(image_arr)
plt.show()

print("Rotated 90_deg")
rotImg=np.rot90(image_arr)
plt.imshow(rotImg)
plt.show()

print("Flipped")
flipImg=np.fliplr(image_arr)
plt.imshow(flipImg)
plt.show()

print("Grayscale")
gray_img=np.dot(image_arr[...,:3],[0.299,0.587,0.114])
plt.imshow(gray_img)

plt.show()
```

```python
from sklearn.datasets import load_iris
import numpy as np
# Load the Iris dataset
iris = load_iris()
# Accessing the features (data) using NumPy array
X = np.array(iris.data) # (Features (sepal length, sepal width, petal length, petal width) #Accessing the target labels (species)
Y = np.array(iris.target) # Target variable (species: 0 for setosa, 1 for versicolor,

feature_names = iris.feature_names
for i in range(X.shape[1]):
    feature = X[:, i]
    print(f"Feature: {feature_names[i]}")
    print(f"  Minimum: {np.min(feature):.2f}")
    print(f"  Maximum: {np.max(feature):.2f}")
    print()


for i in range(X.shape[1]):
  feature=X[:,i]
  print(f"Feature : {feature_names[i]}")
  print(f"Mean :  {np.mean(feature)}")
  print(f"Median :  {np.median(feature)}")
  print(f"Standard Deviation : {np.std(feature)}")

sepal_length_width = X[:, :2]
print("Sepal Length and Sepal Width:")
print(sepal_length_width)




plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=Y, cmap='Purples')
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Sepal Length vs Sepal Width")
plt.colorbar(label="Species")
plt.show()

plt.figure(figsize=(8, 6))
plt.hist(X[:, 0], bins=20, color='skyblue', edgecolor='red')
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.title("Distribution of Sepal Length")
plt.show()


plt.figure(figsize=(8, 6))
plt.plot(X[:, 2], X[:, 3], marker='o', linestyle='-', color='green')
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.title("Petal Length vs Petal Width")
plt.grid(True)
plt.show()
```
