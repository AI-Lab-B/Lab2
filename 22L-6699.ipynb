```python
genomeSequence=[]
file = open("genome.txt")
for line in file:
    for char in line:
        genomeSequence.append(char)
length=len(genomeSequence)

t=np.linspace(0, 4*np.pi, length)
x=np.cos(t)
y=np.sin(t)
z=np.linspace(0,5,length)
coordinates=np.column_stack((x,y,z))

fig=plt.figure()
ax=fig.add_subplot(projection='3d')
colors={'A':'red',
        'T':'green',
        'C':'blue',
        'G':'orange'}
for i,char in enumerate(genomeSequence):
    ax.scatter(coordinates[i,0],coordinates[i,1],coordinates[i,2],color=colors[char],marker='o')
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Genome Visualization")
plt.show()
```

```python
img = mpimg.imread('R.png')
gray_img = np.dot(img[...,:3], [0.2989, 0.5870, 0.1140])
rotated_img=np.rot90(img)
flipped_img=np.fliplr(img)
fig=plt.figure(1,figsize=(20,20))
plt.subplot(1,4,1)
plt.title("Original Image")
plt.imshow(img)
#plot rotated image
plt.subplot(1,4,2)
plt.title("Rotated Image")
plt.imshow(rotated_img)
#plot flipped image
plt.subplot(1,4,3)
plt.title("Flipped Image")
plt.imshow(flipped_img)
#plot grayscale image
plt.subplot(1,4,4)
plt.title("Grayscale Image")
plt.imshow(gray_img,cmap='gray')
plt.show()
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
X=np.array(iris.data)
Y=np.array(iris.target)
mean=np.mean(X,axis=0)
median=np.median(X,axis=0)
std=np.std(X,axis=0)
min_values=np.min(X,axis=0)
max_values=np.max(X,axis=0)
print("Mean:",mean)
print("Median:",median)
print("Standard Deviation:",std)
print("Minimum Values:",min_values)
print("Maximum Values:",max_values)

sepal_length=X[:,0]
sepal_width=X[:,1]
print("Sepal Length:",sepal_length)
print("Sepal Width:",sepal_width)
fig=plt.figure(1,figsize=(20,5))

plt.subplot(1,3,1)
plt.scatter(sepal_length,sepal_width)
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Scatter Plot of Sepal Length vs Sepal Width")

plt.subplot(1,3,2)
plt.hist(sepal_length,bins=10)
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.title("Histogram of Sepal Length")

plt.subplot(1,3,3)
petal_length=X[:,2]
petal_width=X[:,3]
plt.plot(petal_length,petal_width)
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")


plt.show()
```
