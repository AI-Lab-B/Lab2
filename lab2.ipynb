```python
with open('genome.txt','r') as file1:
    data = file1.read()
    genome_seq=data.split('\n')[0]
genome_seq
```

```python
genome_len=len(genome_seq)
genome_len
```

```python
genome_seq=list(genome_seq)
genome_seq
```

```python
t = np.linspace(0, 4 * np.pi, genome_len)
x = np.cos(t)
y = np.sin(t)
z = np.linspace(0, 5, genome_len)
```

```python
coordinates = np.column_stack((x, y, z))
```

```python
colors = {'A': 'red', 'T': 'blue', 'C': 'green', 'G': 'orange'}
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
for i, nucleotide in enumerate(genome_seq):
 ax.scatter(coordinates[i, 0], coordinates[i, 1], coordinates[i, 2],
 color=colors[nucleotide], marker='o')
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Genome Visualization")
plt.show()
```

```python
img= Image.open('img.png')
img
```

```python
npdata=np.asarray(img)
npdata
```

```python
fig = plt.figure()
plot = fig.add_subplot(projection='3d')
plot.set_title('image')
plot.scatter(npdata[:,:,0],npdata[:,:,1],npdata[:,:,2])
plt.show()
```

```python
npdatarotted=np.rot90(npdata)
fig = plt.figure()
plot = fig.add_subplot(projection='3d')
plot.set_title('image after rot')
plot.scatter(npdatarotted[:,:,0],npdatarotted[:,:,1],npdatarotted[:,:,2])
plt.show()
```

```python
plt.imshow(npdatarotted)
plt.show()
```

```python
npdataflipped=np.fliplr(npdata)
fig = plt.figure()
plot2 = fig.add_subplot(projection='3d')
plot2.set_title('image after flip')
plot2.scatter(npdataflipped[:,:,0],npdataflipped[:,:,1],npdataflipped[:,:,2])
plt.show()
```

```python
plt.imshow(npdataflipped)
plt.show()
```

```python
gray_img = np.dot (npdata[..., :3], [0.299, 0.587, 0.114])
plt.imshow(gray_img, cmap='gray')
plt.show()
```

```python
from sklearn.datasets import load_iris
import numpy as np
# Load the Iris dataset
iris = load_iris()
# Accessing the features (data) using NumPy array
X = np.array(iris.data) # (Features (sepal length, sepal width, petal length, petal width) #Accessing the target labels (species)
Y = np.array(iris.target) # Target variable (species: 0 for setosa, 1 for versicolor, 2 for virginica)
```

```python
print(f'mean of sepal length: {np.mean(X[:,0])}')
print(f'mean of sepal width: {np.mean(X[:,1])}')
print(f'mean of petal length: {np.mean(X[:,2])}')
print(f'mean of petal width: {np.mean(X[:,3])}')
```

```python
print(f'median of sepal length: {np.median(X[:,0])}')
print(f'median of sepal width: {np.median(X[:,1])}')
print(f'median of petal length: {np.median(X[:,2])}')
print(f'median of petal width: {np.median(X[:,3])}')
```

```python
print(f'standard deviation of sepal length: {np.std(X[:,0])}')
print(f'standard deviation of sepal width: {np.std(X[:,1])}')
print(f'standard deviation of petal length: {np.std(X[:,2])}')
print(f'standard deviation of petal width: {np.std(X[:,3])}')
```

```python
print(f'max of sepal length: {np.max(X[:,0])}')
print(f'min of sepal length:{np.min(X[:,0])}')
```

```python
print(f'max of sepal width: {np.max(X[:,1])}')
print(f'min of sepal width:{np.min(X[:,1])}')
```

```python
print(f'max of petal length: {np.max(X[:,2])}')
print(f'min of petal length:{np.min(X[:,2])}')
```

```python
print(f'max of petal width: {np.max(X[:,3])}')
print(f'min of petal width:{np.min(X[:,3])}')
```

```python
sepal_length = np.array(X[:,0])
sepal_width = np.array(X[:,1])
```

```python
plt.scatter(sepal_length,sepal_width)
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.show()
```

```python
plt.hist(sepal_length)
plt.xlabel('sepal length')
plt.ylabel('frequency')
plt.show()
```

```python
petal_length = X[:, 2]
petal_width = X[:, 3]

plt.scatter(petal_length, petal_width)
plt.xlabel('Petal Length ')
plt.ylabel('Petal Width ')
plt.title('Petal Length vs. Petal Width')
z = np.polyfit(petal_length, petal_width, 1)
p = np.poly1d(z)
plt.plot(petal_length ,p(petal_length),"r--")
plt.grid(True)
plt.show()
```
