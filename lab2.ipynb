```python
import numpy as np
import matplotlib.pyplot as mpl
import random
file=open('genome.txt','r')
l=[]
for line in file:
    l=line.strip()
file.close()
length=len(l)

# We'll use the parametric equations for a helix:
# x = cos(t), y = sin(t), z = t (or a scaled version of t)
# We want to span a range so that the helix makes a few turns.
t = np.linspace(0, 4 * np.pi, length)  # 4*pi gives about 2 turns
x = np.cos(t)
y = np.sin(t)
z = np.linspace(0, 5, length)  # z increases linearly to spread out the helix vertically
# Combine the coordinates into a (genome_length x 3) array
coordinates = np.column_stack((x, y, z))


# Generate random genome sequence
nucleotides = ["A", "T", "C", "G"]
genome_length = 20
genome = ""
for _ in range(genome_length):
    genome += random.choice(nucleotides)

# Create 3D plot
fig = mpl.figure()
ax = fig.add_subplot(projection='3d')

# Generate random coordinates for each nucleotide
coordinates = np.random.rand(genome_length, 3)

# Color mapping
colors = {'A': 'red', 'T': 'blue', 'C': 'green', 'G': 'orange'}
# Plot each nucleotide with its corresponding color and coordinates
for i, nucleotide in enumerate(genome):
  ax.scatter(coordinates[i, 0], coordinates[i, 1], coordinates[i, 2],  color=colors[nucleotide], marker='o')
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Genome Visualization")
mpl.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image  # Import the PIL library

# 1. Load an image from the internet (or a local file) and convert it to a NumPy array
def load_image_to_numpy_array(image_path):
    """Loads an image from a file and converts it to a NumPy array."""
    try:
        img = Image.open(image_path) # Use PIL to open the image
        img_array = np.array(img)  # Convert to NumPy array
        return img_array
    except FileNotFoundError:
        print(f"Error: The image file '{image_path}' was not found.")
        return None  # Or raise the exception, depending on how you want to handle it
    except Exception as e:
        print(f"Error loading image: {e}")
        return None

# 2. Plot the NumPy array using Matplotlib
def plot_numpy_array_image(img_array, title="Image"):
    """Plots a NumPy array as an image using Matplotlib."""
    if img_array is not None:
        plt.imshow(img_array)
        plt.title(title)
        plt.axis('off')  # Turn off axis labels and ticks
        plt.show()
    else:
        print("No image data to plot.")

# 3. Rotate and flip the image using NumPy
def rotate_and_flip_image(img_array):
    """Rotates and flips the image using NumPy."""
    if img_array is not None:
        rotated_img = np.rot90(img_array)  # Rotate 90 degrees counterclockwise
        flipped_img = np.fliplr(img_array)  # Flip horizontally
        return rotated_img, flipped_img
    else:
        return None, None

# 4. Apply a grayscale filter to the original NumPy image
def apply_grayscale_filter(img_array):
    """Applies a grayscale filter to the original NumPy image."""
    if img_array is not None:
        # Grayscale conversion formula: Y = 0.299*R + 0.587*G + 0.114*B
        gray_img = np.dot(img_array[..., :3], [0.299, 0.587, 0.114])
        # Convert to uint8 for proper display (important for grayscale images)
        gray_img = gray_img.astype(np.uint8)
        return gray_img
    else:
        return None

# Main execution
if __name__ == "__main__":
    image_path = "doggy.webp"  # Replace with the actual path to your image

    # 1. Load the image
    img_array = load_image_to_numpy_array(image_path)

    if img_array is not None:
        # 2. Plot the original image
        plot_numpy_array_image(img_array, title="Original Image")

        # 3. Rotate and flip the image
        rotated_img, flipped_img = rotate_and_flip_image(img_array)

        if rotated_img is not None and flipped_img is not None:
            plot_numpy_array_image(rotated_img, title="Rotated Image")
            plot_numpy_array_image(flipped_img, title="Flipped Image")

        # 4. Apply the grayscale filter
        grayscale_img = apply_grayscale_filter(img_array)

        if grayscale_img is not None:
            plot_numpy_array_image(grayscale_img, title="Grayscale Image")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# 1. Load the Iris dataset
iris = load_iris()
X = np.array(iris.data)  # Features (sepal length, sepal width, petal length, petal width)
Y = np.array(iris.target)  # Target variable (species: 0 for setosa, 1 for versicolor, 2 for virginica)

# 2. NumPy Calculations
# Calculate mean, median, and standard deviation for each feature
feature_names = iris.feature_names  # Get feature names
print("NumPy Calculations:")
for i in range(X.shape[1]):  # Iterate through each column (feature)
    feature_data = X[:, i]  # Extract data for the current feature
    mean_value = np.mean(feature_data)
    median_value = np.median(feature_data)
    std_dev_value = np.std(feature_data)
    min_value = np.min(feature_data)
    max_value = np.max(feature_data)

    print(f"Feature: {feature_names[i]}")
    print(f"  Mean: {mean_value:.2f}")  # Format to 2 decimal places
    print(f"  Median: {median_value:.2f}")
    print(f"  Standard Deviation: {std_dev_value:.2f}")
    print(f"  Minimum: {min_value:.2f}")
    print(f"  Maximum: {max_value:.2f}")
    print("-" * 30)

# Extract sepal length and sepal width as a NumPy array
sepal_data = X[:, :2]  # All rows, first two columns (sepal length and sepal width)
print("\nSepal Length and Sepal Width Data:")
print(sepal_data)

# 3. Matplotlib Visualizations
# Create a scatter plot of sepal length vs sepal width
plt.figure(figsize=(8, 6))  # Adjust figure size for better visualization
plt.scatter(X[:, 0], X[:, 1], c=Y)  # Sepal length vs sepal width, color-coded by species
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Sepal Length vs Sepal Width (Color-Coded by Species)")
plt.colorbar(label="Species (0: setosa, 1: versicolor, 2: virginica)")  # Add a colorbar to indicate species
plt.grid(True)
plt.show()

# Plot a histogram showing the distribution of sepal length
plt.figure(figsize=(8, 6))
plt.hist(X[:, 0], bins=20)  # Sepal length distribution, using 20 bins
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.title("Distribution of Sepal Length")
plt.grid(True)
plt.show()

# Create a line plot to visualize the relationship between petal length and petal width
plt.figure(figsize=(8, 6))
plt.plot(X[:, 2], X[:, 3], marker='o')  # Petal length vs petal width, with markers
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.title("Relationship between Petal Length and Petal Width")
plt.grid(True)
plt.show()
```
