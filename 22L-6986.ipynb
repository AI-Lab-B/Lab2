```python
from pickle import REDUCE
import random
file = open("genome.txt", "r")
content = file.read()
print(content)
lis = list(content)
print(lis)
print(len(content))
genome = ""
genome_length = len(content)
t = np.linspace(0, 4 * np.pi, genome_length)
x = np.cos(t)
y = np.sin(t)
z = np.linspace(0, 5, genome_length)
coordinates = np.column_stack((x, y, z))
for _ in range(genome_length):
 genome += random.choice(lis)
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
coordinates = np.random.rand(genome_length, 3)

colors = {'A': 'red', 'T': 'blue', 'C': 'green', 'G': 'orange'}
for i, lis in enumerate(genome):
 ax.scatter(coordinates[i, 0], coordinates[i, 1], coordinates[i, 2],
 color=colors[lis], marker='o')
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Genome Visualization")
plt.show()
file.close()
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

img = Image.open('xyz.jpg')
numpydata = np.asarray(img)

plt.figure(figsize=(6, 6))
plt.title("Original Image")
plt.imshow(numpydata)
plt.axis("off")
plt.show()

rotated_img = np.rot90(numpydata)

plt.figure(figsize=(6, 6))
plt.title("Rotated Image")
plt.imshow(rotated_img)
plt.axis("off")
plt.show()

flipped_img = np.fliplr(numpydata)

plt.figure(figsize=(6, 6))
plt.title("Flipped Image")
plt.imshow(flipped_img)
plt.axis("off")
plt.show()

gray_img = np.dot(numpydata[..., :3], [0.299, 0.587, 0.114])

plt.figure(figsize=(6, 6))
plt.title("Grayscale Image")
plt.imshow(gray_img, cmap="gray")
plt.axis("off")
plt.show()
```

```python
from sklearn.datasets import load_iris
import numpy as np

iris = load_iris()

X = np.array(iris.data)
Y = np.array(iris.target)

mean_values = np.mean(X, axis=0)
median_values = np.median(X, axis=0)
std_dev_values = np.std(X, axis=0)

min_values = np.min(X, axis=0)
max_values = np.max(X, axis=0)

sepal_features = X[:, :2]
print("Mean values for each feature:", mean_values)
print("Median values for each feature:", median_values)
print("Standard deviation for each feature:", std_dev_values)
print("Minimum values for each feature:", min_values)
print("Maximum values for each feature:", max_values)
print("Sepal features (sepal length and sepal width):\n", sepal_features)

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=Y, cmap='viridis', edgecolor='k')
plt.title("Scatter Plot: Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.colorbar(label="Species")
plt.show()

plt.figure(figsize=(8, 6))
plt.hist(X[:, 0], bins=20, color='skyblue', edgecolor='black')
plt.title("Histogram: Distribution of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(8, 6))
plt.plot(X[:, 2], X[:, 3], 'o-', color='red', label="Petal Length vs Petal Width")
plt.title("Line Plot: Petal Length vs Petal Width")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.legend()
plt.show()
```
