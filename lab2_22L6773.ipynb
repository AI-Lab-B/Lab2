```python
#q2
import numpy as np
import matplotlib.pyplot as plt



f = open("genome.txt", "r")
genome_sequence = f.read()
f.close()

genome_list = list(genome_sequence)
genome_length = len(genome_list)
print(genome_sequence)
print(genome_length)

t = np.linspace(0, 4 * np.pi, genome_length)
x = np.cos(t)
y = np.sin(t)
z = np.linspace(0, 5, genome_length)
coordinates = np.column_stack((x, y, z))


color_map = {
    'A': 'red',
    'T': 'blue',
    'G': 'green',
    'C': 'yellow'
}

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

i=0
for x,y,z in coordinates:
    ax.scatter(x, y, z, color=color_map[genome_list[i]],s=100)
    i+=1

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('Helix Structure of Genome Sequence')
plt.show()





```

```python
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
# Accessing the features (data) using NumPy array
X = np.array(iris.data) # (Features (sepal length, sepal width, petal length, petalwidth) #Accessing the target labels (species)
Y = np.array(iris.target) # Target variable (species: 0 for setosa, 1 for versicolor, 2 for virginica)

feature_names = iris.feature_names
for i in range(X.shape[1]):
    feature = X[:, i]
    print(f"Feature: {feature_names[i]}")
    print(f"  Minimum: {np.min(feature):.2f}")
    print(f"  Maximum: {np.max(feature):.2f}")
    print()

for i in range(X.shape[1]):
    feature = X[:, i]
    print(f"Feature: {feature_names[i]}")
    print(f"  Mean: {np.mean(feature)}")
    print(f"  Median: {np.median(feature)}")
    print(f"  Standard Deviation: {np.std(feature)}")
    print()

sepal_length_width = X[:, :2]
print("Sepal Length and Sepal Width:")
print(sepal_length_width)




plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=Y, cmap='viridis')
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Sepal Length vs Sepal Width")
plt.colorbar(label="Species")
plt.show()

plt.figure(figsize=(8, 6))
plt.hist(X[:, 0], bins=20, color='skyblue', edgecolor='red')
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.title("Distribution of Sepal Length")
plt.show()


plt.figure(figsize=(8, 6))
plt.plot(X[:, 2], X[:, 3], marker='o', linestyle='-', color='pink')
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.title("Petal Length vs Petal Width")
plt.grid(True)
plt.show()
```
