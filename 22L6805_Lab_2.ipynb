```python
# Question 2:

import numpy as np 

with open("genome.txt", 'r') as genome_data:
  genome_sequence = genome_data.read()

genome_list = list(genome_sequence.strip())
genome_length = len(genome_list)

t = np.linspace(0, 4 * np.pi, genome_length)
x = np.cos(t) 
y = np.sin(t) 
z = np.linspace(0, 5, genome_length) 
coordinates = np.column_stack((x, y, z))

fig = plt.figure()
axes = fig.add_subplot(projection='3d')
colors = {'A': 'red', 'T': 'blue', 'C': 'green', 'G': 'orange'}

for i, nucleotide in enumerate(genome_list): 
  axes.scatter(coordinates[i, 0], coordinates[i, 1], coordinates[i, 2], color=colors[nucleotide], s=100)

axes.set_xlabel("X") 
axes.set_ylabel("Y") 
axes.set_zlabel("Z") 
axes.set_title("3D Genome Visualization on a Helix") 
plt.show()
```

```python
# Question 3:

from PIL import Image

img = Image.open("image.jpeg")
img_array = np.asarray(img)

plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.imshow(img_array)
plt.title("Original")
plt.axis("off")

rotated_img = np.rot90(img_array)
plt.subplot(2, 2, 2)
plt.imshow(rotated_img)
plt.title("90 Degree Rotated")
plt.axis("off")

flipped_img = np.fliplr(img_array)
plt.subplot(2, 2, 3)
plt.imshow(flipped_img)
plt.title("Horizontally Flipped")
plt.axis("off")


gray_img = np.dot(img_array[..., :3], [0.299, 0.587, 0.114])
plt.subplot(2, 2, 4)
plt.imshow(gray_img, cmap="gray")
plt.title("Grayscale")
plt.axis("off")

plt.show()
```

```python
# Question 4:

from sklearn.datasets import load_iris

iris = load_iris()

X = np.array(iris.data) 
Y = np.array(iris.target)  

mean_values = np.mean(X, axis=0)
median_values = np.median(X, axis=0)
std_values = np.std(X, axis=0)

min_values = np.min(X, axis=0)
max_values = np.max(X, axis=0)

print(f"Mean values: {mean_values}")
print(f"Median values: {median_values}")
print(f"Standard deviation: {std_values}")
print(f"Minimum values: {min_values}")
print(f"Maximum values: {max_values}")

sepal_features = X[:, :2] 
petal_features = X[:, 2:]

plt.figure(figsize=(8, 5))
plt.scatter(X[:, 0], X[:, 1], c=Y, cmap='viridis', edgecolor='k')
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Scatter Plot of Sepal Length vs Sepal Width")
plt.colorbar(label="Species (0: Setosa, 1: Versicolor, 2: Virginica)")
plt.show()

plt.figure(figsize=(8, 5))
plt.hist(X[:, 0], bins=20, color='skyblue', edgecolor='black')
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.title("Histogram of Sepal Length")
plt.show()

plt.figure(figsize=(8, 5))
plt.plot(petal_features[:, 0], petal_features[:, 1], marker='o', linestyle='-')
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.title("Line Plot of Petal Length vs Petal Width")

# Show Plot
plt.show()
```
