```python
import numpy as np
import matplotlib.pyplot as plt

# We'll use the parametric equations for a helix:
# x = cos(t), y = sin(t), z = t (or a scaled version of t)
# We want to span a range so that the helix makes a few turns.
with open("genome.txt", 'r') as file:
  f = file.read().strip()
g_seq = list(f)
print(g_seq)
genome_length = len(g_seq)
print(genome_length)

# Color dictionary to match genome with color
COLORS = {
  "A": 'Red',
  "T": 'Yellow',
  "C": 'Blue',
  "G": 'Green'
}

cols = [COLORS.get(color) for color in g_seq]
print(cols)

t = np.linspace(0, 4 * np.pi, genome_length) # 4*pi gives about 2 turns
x = np.cos(t)
y = np.sin(t)
z = np.linspace(0, 5, genome_length) # z increases linearly to spread out the helix vertically
# Combine the coordinates into a (genome_length x 3) array
coordinates = np.column_stack((x, y, z))


# plotting
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(coordinates[:,0], coordinates[:,1], coordinates[:,2],c=cols,s=30)
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

fig.suptitle("3D Genome Visualization on a Helix ")
# ax.plot3D(x,y,z,cols)
plt.show()



# listOfColors=[colors.get(letter) for letter in g_seq]
# fig = plt.figure()
# ax = fig.add_subplot( projection='3d')
# ax.scatter(coordinates[:, 0], coordinates[:, 1], coordinates[:, 2],
#            c=listOfColors, s=50)  # s controls the marker size

# # Set the axis labels and title.
# ax.set_xlabel("X")
# ax.set_ylabel("Y")
# ax.set_zlabel("Z")
# ax.set_title("3D Helix Structure with Color-coded Molecules")

# plt.show()


# with open("genome.txt", "r") as file:
#     hehe = file.read().strip()

# g_seq=list(hehe)
# print(g_seq)
# genome_length = len(g_seq)
# print("Genome length:", genome_length)

# t = np.linspace(0, 4 * np.pi, genome_length)
# x = np.cos(t)                    # x = cos(t)
# y = np.sin(t)                    # y = sin(t)
# z = np.linspace(0, 5, genome_length)  # z increases linearly to spread out the helix

# coordinates = np.column_stack((x, y, z))

# colors = {
#     'A': 'red',
#     'T': 'blue',
#     'C': 'green',
#     'G': 'orange'
# }

```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

img_path = "image.jpg"

img = Image.open(img_path)
img_array = np.array(img)

rotated = np.rot90(img_array)
flipped = np.fliplr(img_array)
gray_img = np.dot (img_array[..., :3], [0.299, 0.587, 0.114])

fig, images = plt.subplots(1, 4, figsize=(17, 5))
images[0].imshow(img_array)
images[0].set_title("Original Image")

images[1].imshow(rotated)
images[1].set_title("Rotated")

images[2].imshow(flipped)
images[2].set_title("Flipped Image")

images[3].imshow(gray_img)
images[3].set_title("Gray Scale Image")


plt.show()

```

```python
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt
# Load the Iris dataset
iris = load_iris()
# Accessing the features (data) using NumPy array
X = np.array(iris.data) # (Features (sepal length, sepal width, petal length, petal width) #Accessing the target labels (species)
Y = np.array(iris.target) # Target variable (species: 0 for setosa, 1 for versicolor, 2 for virginica)

means = np.mean(X, axis=0)
medians = np.median(X, axis=0)
stds = np.std(X, axis=0)
minimums = np.min(X, axis=0)
maximums = np.max(X, axis=0)

print("Means: ",means)
print("Medians: ",medians)
print("Standard Deviations: ", stds)
print("Minimum Values: ", minimums)
print("Maximum Values: ", maximums)

sepal_len = X[:, 0]
sepal_width = X[:, 1]

# Scatter plot: Sepal Length vs Sepal Width
plt.figure(figsize=(7,5))
plt.scatter(sepal_len, sepal_width, c=Y, cmap='viridis', edgecolor='k', alpha=0.7)
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Scatter Plot: Sepal Length vs Sepal Width")
plt.colorbar(label="Species")
plt.show()

# Histogram of Sepal Length
plt.figure(figsize=(7,5))
plt.hist(sepal_len, bins=20, color='blue', edgecolor='black', alpha=0.7)
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.title("Histogram of Sepal Length")
plt.show()

# Line plot: Petal Length vs Petal Width
petal_len = X[:, 2]
petal_width = X[:, 3]

plt.figure(figsize=(7,5))
plt.plot(petal_len, petal_width, 'o-', color='red', alpha=0.7)
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.title("Line Plot: Petal Length vs Petal Width")
plt.show()
```
