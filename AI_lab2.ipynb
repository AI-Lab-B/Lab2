```python
import matplotlib.pyplot as plt
import numpy as np

def main():
  file = open("genome.txt", "r")
  genomeTxt = (file.read()).strip("\n")

  genome_sequence = list(genomeTxt)

  genome_length = len(genome_sequence)

  # We'll use the parametric equations for a helix:
  # x = cos(t), y = sin(t), z = t (or a scaled version of t)
  # We want to span a range so that the helix makes a few turns.
  t = np.linspace(0, 4 * np.pi, genome_length) # 4*pi gives about 2 turns

  x = np.cos(t)
  y = np.sin(t)
  z = np.linspace(0, 5, genome_length) # z increases linearly to spread out the helix vertically
  # Combine the coordinates into a (genome_length x 3) array
  coordinates = np.column_stack((x, y, z))

  colors = {'A':"red",'T':"blue",'G':"green",'C':"orange"}

  fig = plt.figure(figsize = (10, 7))
  ax = plt.axes(projection ="3d")

  # Creating plot
  i = 0
  for x,y,z in coordinates:
    ax.scatter(x,y,z,c=colors[genome_sequence[i]], s=200)
    i += 1

  plt.title("3D Genome Visualization of a Helix")
  ax.set_xlabel('X')
  ax.set_ylabel('Y')
  ax.set_zlabel('Z')

  # show plot
  plt.show()

main()
```

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

def main():
  im = Image.open("th.jpg")
  imgArr = np.asarray(im)

  plt.imshow(imgArr)
  plt.show()

  flippedImg = np.fliplr(imgArr)
  plt.imshow(flippedImg)
  plt.show()

  rotImg = np.rot90(imgArr)
  plt.imshow(rotImg)
  plt.show()

  gray_img = np.dot(imgArr[..., :3], [0.299, 0.587, 0.114])
  plt.imshow(gray_img)
  plt.show()

main()

```

```python
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()

# Accessing the features (data) using NumPy array
X = np.array(iris.data) # (Features (sepal length, sepal width, petal length, petal width)
#Accessing the target labels (species)
Y = np.array(iris.target) # Target variable (species: 0 for setosa, 1 for versicolor, 2 for virginica)

for i in range(len(X)):
  print(f"Mean: {np.mean[X[i]]}")
  print(f"Median: {np.median[X[i]]}")
  print(f"Standard Deviation: {np.std[X[i]]}")
  print(f"Minimum: {np.min[X[i]]}")
  print(f"Maximum: {np.max[X[i]]}")

sp = [X[:, 0], X[:, 1]]

plt.scatter(sp[0], sp[1])
plt.title("Sepal Length vs Sepal Height Scatter Plot")
plt.xlabel('Length (cm)')
plt.ylabel('Length (cm)')
plt.show()

plt.hist(sp[0], bins=len(sp[0]))
plt.title("Sepal Length Histogram")
plt.xlabel('Length (cm)')
plt.ylabel('Number of times')
plt.show()

plt.hist(sp[1], bins=len(sp[0]))
plt.title("Sepal Height Histogram")
plt.xlabel('Length (cm)')
plt.ylabel('Number of times')
plt.show()

plt.plot(sp[0], sp[1])
plt.show()
```
