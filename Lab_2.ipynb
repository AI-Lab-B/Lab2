```python
import matplotlib.pyplot as plt
import numpy as np

file=open("genome.txt")
genome_sequence=file.read()
genome_sequence=[ch for ch in genome_sequence]

genome_length=len(genome_sequence)
print(genome_length)
t = np.linspace(0, 4 * np.pi, genome_length)
x = np.cos(t)
y = np.sin(t)
z = np.linspace(0, 5, genome_length)

color_map={
    "A":"red",
    "T":"blue",
    "G":"green",
    "C":"yellow"
}

fig= plt.figure()
ax=fig.add_subplot(projection='3d')

coordinates = np.column_stack((x, y, z))
i=0;
for x,y,z in coordinates:
  ax.scatter(x,y,z,c=color_map[genome_sequence[i]],s=100)
  i+=1


ax.set_xlabel("X");
ax.set_ylabel("Y");
plt.title("Helix Strucutre of genome sequence")
plt.show()
```
