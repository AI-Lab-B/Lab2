```python

import matplotlib.pyplot as plt
import numpy as np
import random
# Generate random genome sequence
file = open("genome.txt", "r")
content = file.read()
nucleotides = list(content)
genome_length = len(nucleotides)
t = np.linspace(0, 4 * np.pi, genome_length)
x = np.cos(t)
y = np.sin(t)
z = np.linspace(0, 5, genome_length)
# Combine the coordinates into a (genome_length x 3) array
coordinates = np.column_stack((x, y, z))
genome = ""
for i in range(genome_length):
 genome += nucleotides[i]
# Create 3D plot
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
# Color mapping
colors = {'A': 'red', 'T': 'blue', 'C': 'green', 'G': 'orange'}
# Plot each nucleotide with its corresponding color and coordinates
for i, nucleotide in enumerate(genome):
 ax.scatter(coordinates[i, 0], coordinates[i, 1], coordinates[i, 2],color=colors[nucleotide], marker='o')
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Genome Visualization")
plt.show()

```

```python
from sklearn.datasets import load_iris
import numpy as np
# Load the Iris dataset
iris = load_iris()
# Accessing the features (data) using NumPy array
X = np.array(iris.data) # (Features (sepal length, sepal width, petal length, petalwidth) #Accessing the target labels (species)
Y = np.array(iris.target) # Target variable (species: 0 for setosa, 1 for versicolor, 2for virginica)
temp = X
sepal_length = list(zip(*temp))[0]
sepal_width = list(zip(*temp))[1]
plt.scatter(sepal_length,sepal_width)
plt.show()
plt.hist(sepal_length, bins=10)
plt.show()

x=list(zip(*temp))[2]
y=list(zip(*temp))[3]
plt.plot(x, y)
plt.show()

mean = np.mean(sepal_length)
median = np.median(sepal_length)
std_deviation = np.std(sepal_length)
print("mean=",mean,"\n")
print("median=",median,"\n")
print("dev=",std_deviation,"\n")

mean = np.mean(sepal_width)
median = np.median(sepal_width)
std_deviation = np.std(sepal_width)
print("mean=",mean,"\n")
print("median=",median,"\n")
print("dev=",std_deviation,"\n")

mean = np.mean(x)
median = np.median(x)
std_deviation = np.std(x)
print("mean=",mean,"\n")
print("median=",median,"\n")
print("dev=",std_deviation,"\n")

mean = np.mean(y)
median = np.median(y)
std_deviation = np.std(y)
print("mean=",mean,"\n")
print("median=",median,"\n")
print("dev=",std_deviation,"\n")
```
